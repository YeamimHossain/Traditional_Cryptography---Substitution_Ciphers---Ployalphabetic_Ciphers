<h1 style = 'text-align: center'><b>Playfair Cipher</b></h1>
<h3>
    <p><b>The playfair cipher uses three rules for encryption:</b>
        <ol>
            <li>If the two letters in a pair are located in the same row of the secret key matrix, the corresponding encrypted character for each letter is the next letter to the right  in the same row (with wrapping to the beginning of the row if the plaintext letter is the last character in the row).</li>
            <br>
            <li>If the two letters in a pair are located in the same column of the secret key matrix, the corresponding encrypted character for each letter is the letter beneath it in the same column (with wrapping to the beginning of the column if the plaintext letter is the last character in the column).</li>
            <br>
            <li>If the two letters in a pair are not located in the same row or column of the secret key matrix, the corresponding encrypted character for each letter is a letter that is in its own row but in the same column as the other letter.</li>
        </ol>
    </p>
    <p><b>The playfair cipher uses three rules for decryption:</b>
        <ol>
            <li>If the two ciphertext letters in a pair are located in the same row of the secret key matrix, the corresponding decrypted character for each letter is the previous letter to the left in the same row (with wrapping to the end of the row if the ciphertext letter is the first character in the row).</li>
            <br>
            <li>If the two ciphertext letters in a pair are located in the same column of the secret key matrix, the corresponding decrypted character for each letter is the letter above it in the same column (with wrapping to the end of the column if the ciphertext letter is the first character in the column).</li>
            <br>
            <li>If the two ciphertext letters in a pair are not located in the same row or column of the secret key matrix, the corresponding decrypted character for each letter is a letter that is in its own row but in the same column as the other letter.</li>
        </ol>
    </p>
</h3>

<h3>Necessary Functions</h3>

```python
# Finding row and column values from the secret_key for the message characters
def rc_val(character, seckey_mat):
    '''
    rc_val returns row and column values of a message character from the secret-key matrix.
    character = character can be any letter in range a-z
    seckey_mat = A specific combination of 5*5 secrect key matrix which contains 26 English letters (a-z) 
    
    # Example
    # rowcol_val = rc_val('m', seckey_mat)
    # print(rowcol_val)
    
    # Output:
    [2, 3]
    Here, it represents the position of the character 'm' in the seckey_mat where 2 represents the row number of the character 
    and 3 represents the column number of the character from the secrect-key matrix
    '''
    rowcol_val = []
    i = 0
    while i < len(seckey_mat):
        j = 0
        while j < len(seckey_mat[i]):
            if len(seckey_mat[i][j]) == 1:
                if seckey_mat[i][j] == character:
                    rowcol_val.append(i)
                    rowcol_val.append(j)
                    break

            elif len(seckey_mat[i][j]) == 2:                
                k = 0
                while k < len(seckey_mat[i][j]):
                    if seckey_mat[i][j][k] == character:
                        rowcol_val.append(i)
                        rowcol_val.append(j)
                        rowcol_val.append([k])
                    k+=1
            j+=1
        i+=1
    return rowcol_val
```

```python
# Message filtering and cleaning function
def prep_msg(userin):
    '''
    # The 'prep_msg' function returns 'mtext' which is the filtered data of 'userin' 
    where space or other sign will be removed and the capital letters will be coverted into small letter.  
    # userin = user input message or string
    # Modulus = The set of list of the total character for the algorithm

    # Example
    exp = prep_msg("I am a boy", 26)
    print(exp)

    Output:
    ['i', 'a', 'm', 'a', 'b', 'o', 'y']
    '''
    # Generating Small letters (English alphabets 'a-z')
    SL = []
    for x in range(ord('a'),ord('z')+1):
        SL.append(chr(x))
    # print(SL)
    
    # Generating Capital letters (English alphabets 'A-Z')
    CL = []
    for x in range(ord('A'),ord('Z')+1):
        CL.append(chr(x))
    # print(CL)
    
    # PlainText and CipherText
    PT = SL
    CT = CL
    
    
    # List of input characters
    inlist = list(userin)
    is_inlist = 0
    while is_inlist<len(inlist):
        if inlist[is_inlist] in CT:
            inlist[is_inlist] = inlist[is_inlist].lower()
        is_inlist+=1
    # print(inlist)
    
    # Removing blank spaces from the message 
    mtext = []      # mtext is the message text or PlainText including only alphabets
    for x in inlist:
        if x in PT:
            mtext.append(x)
    # print(mtext)
    return mtext
```

```python
# Finding specific character value of PT or CT
def PTCT_value(character, PT):
    i = 0
    while i<len(PT):
        if character == PT[i]:
            pos = i
        i+=1
    # print(f'The value of {character} is {pos}.')
    return pos
```

<h3>3. Encrypting the Message</h3>

```python
# Filtering the message
mtext = prep_msg(userin)
print(mtext)
```

```python
# Finding the closet suitable length for encryption
pair = 2
len_msg = len(mtext)
while len_msg % pair != 0:
    len_msg += 1
print(len_msg)
```

```python
# Finding the required number of bogus character & generating bogus character
bogus_req = len_msg - len(mtext)        # bogus_req is the total number of required bogus characters
print(f'Bogus character required: {bogus_req}')
bogus = []
bogus_char = Modulus       # bogus_char is the bogus character value
i = 0
while i < bogus_req:
    bogus_char = bogus_char - 1
    bogus.append(PT[bogus_char])
    i+=1
bogus = bogus[::-1]
print(f'Bogus character list: {bogus}')
```

```python
# Adding the bogus character with the message
new_mtext = mtext + bogus
print(f'The updated mesasge: {new_mtext}')
```

```python
# Making the pair from the message
pair_list = []       # pair_list is the list of pair characters of the message
num_block = int(len(new_mtext)/pair)
i = 1
j = 1
while i < num_block + 1:
    sng_pair = []      # sng_pair is the single pair
    while j < pair * i + 1:
        sng_pair.append(new_mtext[j-1])
        j+=1
    pair_list.append(sng_pair)
    i+=1
print(f'Two-character pair list: {pair_list}')
```

<h3>Check and rearrange if the same character exist in any pair</h3>

```python
# Check the last block is the same then add new_block ['y', 'z']
if pair_list[num_block-1][0] == pair_list[num_block-1][1]:
    pair_list.append(['y', 'z'])
# print(pair_list)

# Rearranging the block if same character exists in any block
i = 0
while i < num_block:
    if pair_list[i][0] == pair_list[i][1]:
        pair_list[i][1], pair_list[i+1][0] = pair_list[i+1][0], pair_list[i][1]
    i+=1
print(pair_list)
```

```python
# Function for encrypt the message using PlayFair algorithm 
def pf_encrypt(pair_list, seckey_mat):
    '''
    It will return CipherText of the input message.
    
    # Example
    # pair_list = [['m', 'e'], ['e', 't'], ['m', 'e']]
    # seckey_mat = [
        ['g','h','b','l','d'],
        ['v','n','o','r','i'],
        ['k','a',['y','z'],'q','s'],
        ['c','x','f','u','m'],
        ['w','p','t','j','e']
    ]
    # cytext = pf_encrypt(pair_list, seckey_mat)
    # print(cytext)

    # Output:
    # [['e', 'd'], ['w', 'j'], ['e', 'd']]
    '''
    import random
    
    cytext = []
    key_len = len(seckey_mat) - 1   # 'key_len' is the length of the secret-key matrix
    
    for x in pair_list:
        # This section will find the row and col values of the message characters from the secrect-key matrix 
        pair_rc = []      # 'pair_rc' will contain the row and column values of the pair
        i = 0
        while i < len(x):
            rowcol_val = rc_val(x[i], seckey_mat)
            pair_rc.append(rowcol_val)
            i+=1

        # If the rows of the character pair are the same
        if pair_rc[0][0] == pair_rc[1][0]:
            if pair_rc[0][1] == key_len and pair_rc[1][1] != key_len:
                pair_rc[0][1] = pair_rc[0][1] - key_len
                pair_rc[1][1] = pair_rc[1][1] + 1
            elif pair_rc[0][1] != key_len and pair_rc[1][1] == key_len:
                pair_rc[0][1] = pair_rc[0][1] + 1
                pair_rc[1][1] = pair_rc[1][1] - key_len
            elif pair_rc[0][1] != key_len and pair_rc[1][1] != key_len:
                pair_rc[0][1] = pair_rc[0][1] + 1
                pair_rc[1][1] = pair_rc[1][1] + 1
        
        # If the columns of the character pair are the same
        elif pair_rc[0][1] == pair_rc[1][1]:
            if pair_rc[0][0] == key_len and pair_rc[1][0] != key_len:
                pair_rc[0][0] = pair_rc[0][0] - key_len
                pair_rc[1][0] = pair_rc[1][0] + 1
            elif pair_rc[0][0] != key_len and pair_rc[1][0] == key_len:
                pair_rc[0][0] = pair_rc[0][0] + 1
                pair_rc[1][0] = pair_rc[1][0] - key_len
            elif pair_rc[0][0] != key_len and pair_rc[1][0] != key_len:
                pair_rc[0][0] = pair_rc[0][0] + 1
                pair_rc[1][0] = pair_rc[1][0] + 1
        
        # If the rows and colimns of the character pair are not the same
        elif pair_rc[0][0] != pair_rc[1][0] or pair_rc[0][1] != pair_rc[1][1]:
            pair_rc[0][1], pair_rc[1][1] = pair_rc[1][1], pair_rc[0][1]

        
        
        # This section will generate the ciphertext based on the row and column positions of the characters
        character = []
        for x in pair_rc:
            if x[0] == 2 and x[1] == 2:
                random.seed(3)
                ran_char = random.randint(0, 1)
                character.append(seckey_mat[x[0]][x[1]][ran_char])
            else:
                character.append(seckey_mat[x[0]][x[1]])
        cytext.append(character)
        
    return cytext
```

```python
# Generating CipherText
cytext = pf_encrypt(pair_list, seckey_PT)
# print(cytext)

# Converting the 2D array to 1D array
mrg_cytext = []
for x in cytext:
    mrg_char = ' '.join(x).replace(' ', '')
    mrg_cytext.append(mrg_char)
mrg_cytext = ' '.join(mrg_cytext).replace(' ', '')
# Converting lowercase value into uppercase value
mrg_cytext = mrg_cytext.upper()
print(f'CiperText: {mrg_cytext}')
```

```python
# Function for decrypt the message using PlayFair algorithm
def pf_decrypt(cypair_list, seckey_mat):
    '''
    It will return PlainText from the CipherText.
    
    # Example
    # cypair_list = [['E', 'D'], ['W', 'J'], ['E', 'D']]
    # seckey_mat = [
        ['G','H','B','L','D'],
        ['V','N','O','R','I'],
        ['K','A',['Y','Z'],'Q','S'],
        ['C','X','F','U','M'],
        ['W','P','T','J','E']
    ]
    # platext = pf_decrypt(cypair_list, seckey_mat)
    # print(platext)

    # Output:
    # [['M', 'E'], ['E', 'T'], ['M', 'E']]
    '''
    import random
    
    platext = []
    key_len = len(seckey_mat) - 1   # 'key_len' is the length of the secret-key matrix
    
    for x in cypair_list:
        # This section will find the row and col values of the ciphertext characters from the secrect-key matrix 
        pair_rc = []      # 'pair_rc' will contain the row and column values of the pair
        i = 0
        while i < len(x):
            rowcol_val = rc_val(x[i], seckey_mat)
            pair_rc.append(rowcol_val)
            i+=1

        # If the rows of the character pair are the same
        if pair_rc[0][0] == pair_rc[1][0]:
            if pair_rc[0][1] == 0 and pair_rc[1][1] != 0:
                pair_rc[0][1] = pair_rc[0][1] + key_len
                pair_rc[1][1] = pair_rc[1][1] - 1
            elif pair_rc[0][1] != 0 and pair_rc[1][1] == 0:
                pair_rc[0][1] = pair_rc[0][1] - 1
                pair_rc[1][1] = pair_rc[1][1] + key_len
            elif pair_rc[0][1] != 0 and pair_rc[1][1] != 0:
                pair_rc[0][1] = pair_rc[0][1] - 1
                pair_rc[1][1] = pair_rc[1][1] - 1
        
        # If the columns of the character pair are the same
        elif pair_rc[0][1] == pair_rc[1][1]:
            if pair_rc[0][0] == 0 and pair_rc[1][0] != 0:
                pair_rc[0][0] = pair_rc[0][0] + key_len
                pair_rc[1][0] = pair_rc[1][0] - 1
            elif pair_rc[0][0] != 0 and pair_rc[1][0] == 0:
                pair_rc[0][0] = pair_rc[0][0] - 1
                pair_rc[1][0] = pair_rc[1][0] + key_len
            elif pair_rc[0][0] != 0 and pair_rc[1][0] != 0:
                pair_rc[0][0] = pair_rc[0][0] - 1
                pair_rc[1][0] = pair_rc[1][0] - 1
        
        # If the rows and colimns of the character pair are not the same
        elif pair_rc[0][0] != pair_rc[1][0] or pair_rc[0][1] != pair_rc[1][1]:
            pair_rc[0][1], pair_rc[1][1] = pair_rc[1][1], pair_rc[0][1]

        
        
        # This section will generate the ciphertext based on the row and column positions of the characters
        character = []
        for x in pair_rc:
            if x[0] == 2 and x[1] == 2:
                random.seed(3)
                ran_char = random.randint(0, 1)
                character.append(seckey_mat[x[0]][x[1]][ran_char])
            else:
                character.append(seckey_mat[x[0]][x[1]])
        platext.append(character)
        
    return platext
```

```python
# Generating PlainText
platext = pf_decrypt(cypair_list, seckey_CT)
# print(platext)

# Converting the 2D array to 1D array
mrg_platext = []
for x in platext:
    mrg_char = ' '.join(x).replace(' ', '')
    mrg_platext.append(mrg_char)
mrg_platext = ' '.join(mrg_platext).replace(' ', '')

# Converting uppercase value into lowercase value
mrg_platext = mrg_platext.lower()
print(f'CiperText: {mrg_platext}')
```
